```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import mnist
from keras.layers import Dense, Input
from keras.models import Model
```

```python
(x, _), (y, _) = mnist.load_data()
```

```python
x = x.astype('float32') / 255
y = y.astype('float32') / 255

x = x.reshape((len(x), np.prod(x.shape[1:])))
y = y.reshape((len(y), np.prod(y.shape[1:])))
```

```python
encDim = 36
ipImg = Input(shape = (784, ))

opDim = 784
```

```python
encoded = Dense(encDim, activation = 'relu')(ipImg)
encoder = Model(ipImg, encoded)
decoded = Dense(opDim, activation = 'sigmoid')(encoded)
```

```python
ae = Model(ipImg, decoded)
ae.compile(optimizer = 'adadelta', loss = 'binary_crossentropy')
```

```python
ae.fit(x, x,  epochs = 15, batch_size = 256, shuffle = True, validation_data = (y, y))
```

```python
decImgs = ae.predict(y)
encImgs = encoder.predict(y)
```
